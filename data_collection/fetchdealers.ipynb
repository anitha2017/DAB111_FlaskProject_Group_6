```python
##Import selenium_webdriver.py python file
import selenium_webdriver as sw
import pandas as pd
import time
```

```python


##chromedriver is required to launch the ford dealership website
chrome_driver_path='.\chromedriver.exe'
ford_url = "https://www.ford.ca/dealerships"
toronto_search_term="Toronto,ON"

##Call method "fetch_dealerlist_webscraping" which is defined in selenium_webdriver.py file 
# pass these parameters driver path, website url, search city as Toronto,ON
dealer_df = sw.fetch_dealerlist_webscraping(webdriver_path=chrome_driver_path, url=ford_url,search_term=toronto_search_term)
```

```python
##list the dealers
dealer_df
```

```python
## fetch website information in a dataframe

dealer_df_new = pd.DataFrame()
dealer_df_new = dealer_df
dealer_df_new['Dealer_Address'] = dealer_df_new['Address'].apply(lambda x: x.split('\n')[0])
dealer_df_new['Dealer_Phone'] = dealer_df_new['Address'].apply(lambda x: x.split('\n')[-1])
dealer_df_new=dealer_df_new.drop(columns="Address")
dealer_df_new.head()
```

```python
##add dummy text when dealer_phone is empty
mask=dealer_df_new['Dealer_Phone'].apply(lambda x: len(x)>15)
dealer_df_new.loc[mask,'Dealer_Phone'] = '(000) 000-0000'
```

```python
##display dummy text
dealer_df_new.loc[mask,'Dealer_Phone'] 
```

```python
##repeat above procedure for windsor dealers list

windsor_url = "https://www.ford.ca/dealerships"
windsor_search_term="Windsor,ON"
time.sleep(2)
##Call method "fetch_dealerlist_webscraping" which is defined in selenium_webdriver.py file 
# pass these parameters driver path, website url, search city as Windsor,ON
windsor_dealer_df = sw.fetch_dealerlist_webscraping(webdriver_path=chrome_driver_path, url=windsor_url,search_term=windsor_search_term)
```

```python
## fetch website information in a dataframe
windsor_dealer_df_new = windsor_dealer_df
windsor_dealer_df_new['Dealer_Address'] = windsor_dealer_df_new['Address'].apply(lambda x: x.split('\n')[0])
windsor_dealer_df_new['Dealer_Phone'] = windsor_dealer_df_new['Address'].apply(lambda x: x.split('\n')[-1])
windsor_dealer_df_new=windsor_dealer_df_new.drop(columns="Address")
windsor_dealer_df_new.head()
```

```python
windsor_dealer_df_new
```

```python
##concatenate toronto and windsor dataframe
df_merged=pd.concat([dealer_df_new,windsor_dealer_df_new], ignore_index=True)
data=pd.DataFrame(df_merged)
```
